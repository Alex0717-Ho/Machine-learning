```
# ----- Input data -----             


from sklearn.datasets import load_iris
import pandas as pd      
iris = load_iris()                 #存成Dict
df = pd.DataFrame(iris["data"],          #原本是Array(字典)-轉
          columns=iris["feature_names"])  #要設定欄位名稱
df["ans"] = iris["target"]             #增加答案欄位
df

# 150row(筆資料) X 5columns(i/o) 
# 0=setosa, 1=versicolor, 2=virginica
# sepal(花萼) petal(花瓣) 
 
# -----  補充  -----
#  machine learning :千-萬筆資料
#  deep   learning :萬-10萬筆資料
#判別困難度與否在 " 欄位多少 = 維度 "                       
```

# 新增區段

```
#----- 評估模型+轉換Numpy+切分資料 ------

                           
import numpy as np                    #model_selection:
from sklearn.model_selection import train_test_split  #評估model 的選擇
#List[題目90, 題目10, 答案90, 答案10]
x_train, x_test, y_train, y_test = train_test_split(np.array(iris["data"]),
                              np.array(iris["target"]),
                              test_size=0.1)

x_train, x_test, y_train, y_test
# 要打才會print
```

```
#----- 訓練模型+Algorithm  ------


from sklearn.tree import DecisionTreeClassifier #Al:DecisionTree

clf = DecisionTreeClassifier(max_depth=2)
# 設定最大深度到2層

# 重要函式fit():給他資料去學習
clf.fit(x_train, y_train)

#執行完就會得到完整的公式
```

```
#----- 資料視覺化 ------


import matplotlib.pyplot as plt  #類似Matlab用來畫圖

from sklearn.tree import plot_tree #畫圖
plt.figure(figsize=(14, 14))     #設定畫布大小(整張圖)

plot_tree(clf, 
     feature_names=iris["feature_names"],  #欄位名稱
     class_names=iris["target_names"],    #答案
     filled=True)               #背景上色
plot_tree
```

```
#----- 驗證 ------     計算機率


from sklearn.metrics import accuracy_score    #所有的度量標準_metrics 
pre = clf.predict(x_test)             #正確率:accuracy_score
accuracy_score(pre, y_test)
```
